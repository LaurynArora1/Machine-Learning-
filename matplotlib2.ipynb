```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python

x=np.linspace(0,5,11)
y=x**2

```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',lw=3,alpha=0.75,ls='--',marker='s',markersize=10,markerfacecolor='orange',markeredgewidth=3,markeredgecolor='green')
#colour='FF8C00', RGB Hex Code
#lw=linewidth
#ls=linestyle
axes.set_xlabel('X-axis')
axes.set_ylabel('Y-axis')
axes.set_title('Title is here')
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlim([0,1])
axes.set_ylim([0,2])
```
