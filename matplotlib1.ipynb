```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL
plt.plot(x,y)
#plt.show()
```

```python
plt.plot(x,y,'r-')
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('I put this title')
```

```python
plt.subplot(1,2,2)
# subplot (no. of rows,no. of coloumns,plot number)
plt.plot(x,y,'r-')

plt.subplot(1,2,1)
# subplot (no. of rows,no. of coloumns,plot number)
plt.plot(y,x,'b')
```

```python
# object oriented method

fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X-axis')
axes.set_ylabel('Y-axis')
axes.set_title('title is here')
```

```python

fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.45,0.2,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x,'r-')
```
