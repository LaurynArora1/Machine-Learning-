```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
flights=sns.load_dataset('flights')
```

```python
flights
```

```python
t=tips.corr()
```

```python
sns.heatmap(t,annot=True)
```

```python
t
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidths=2)
```

```python
fp
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=2)
# clusters similar data together
```
