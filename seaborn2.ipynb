CATEGORICAL PLOTS

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')

first_2rows=tips.head(2)
tips.head()
# head() is used for getting the first n rows
```

```python
first_2rows
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
# barplot helps us aggregate categorical data based of some function(mean,std. deviation)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
# swarmplot is a combination of a stripplot and a violinplot
#it is not usually recommended to use this
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='purple')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
# factorplot has been renamed to catplot
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```
